```python
from pytriqs.gf import *
from pytriqs.archive import *
from pytriqs.plot.mpl_interface import oplot, oploti, oplotr, plt
import numpy as np
plt.rcParams["figure.figsize"] = (14,9)

pom = HDFArchive('results/pomerol.h5','r')

color=['grey', 'red', 'green', 'blue', 'black']
#titlecolor=['grey', 'red', 'green', 'black']

plt.rcParams.update({'font.size': 10})

for m in {"I", "R"}:
    for s in {"up", "dn"}:
        for b in range(0,2):

            data = []

            w2dyn_vars = []
            triqs_vars = []

            times = [500, 1000, 2000]

            for nt, max_time in enumerate(times):

                folder = "results_"+str(max_time)+"/"
                #print 'folder', folder

                w2dyn = HDFArchive(folder+'w2dyn_cthyb.h5','r')
                triqs = HDFArchive(folder+'cthyb.h5','r')

                beta = w2dyn["G/dn/mesh/domain/beta"]
                #print "beta", beta
                iw = triqs["Solver_Info/solver/G0_iw/up/mesh"]

                iw_max = 30

                iw_ = []
                ind_min = 10000
                ind_max = 0
                for n,i in enumerate(iw):
                    #print 'i', np.imag(i.value)
                    #print 'n,i', n,i
                    iw_ = np.append(iw_, np.imag(i.value))

                    if np.imag(i.value) < -iw_max:
                        ind_min = n
                ind_max = iw_.shape[0] - ind_min - 1

                #print 'iw_', iw_
                #print 'ind_min', ind_min
                #print 'ind_max', ind_max
                #print 'iw_[ind_min]', iw_[ind_min]
                #print 'iw_[ind_max]', iw_[ind_max]

                max_triqs = np.amax(np.real(triqs['G'][s][b,b].data[ind_min:ind_max]-pom['G'][s][b,b].data[ind_min:ind_max]))
                max_w2dyn = np.amax(np.real(w2dyn['G'][s][b,b].data[ind_min:ind_max]-pom['G'][s][b,b].data[ind_min:ind_max]))

                min_triqs = np.amin(np.real(triqs['G'][s][b,b].data[ind_min:ind_max]-pom['G'][s][b,b].data[ind_min:ind_max]))
                min_w2dyn = np.amin(np.real(w2dyn['G'][s][b,b].data[ind_min:ind_max]-pom['G'][s][b,b].data[ind_min:ind_max]))

                ymin = min(min_triqs, min_w2dyn)
                ymax = max(max_triqs, max_w2dyn)

                ### calculate variance

                w2dyn_var = np.var(np.real(w2dyn['G'][s][b,b].data[ind_min:ind_max]-pom['G'][s][b,b].data[ind_min:ind_max]))
                triqs_var = np.var(np.real(triqs['G'][s][b,b].data[ind_min:ind_max]-pom['G'][s][b,b].data[ind_min:ind_max]))
                #print 'w2dyn_var', w2dyn_var
                #print 'triqs_var', triqs_var

                w2dyn_vars = np.append(w2dyn_vars, w2dyn_var)
                triqs_vars = np.append(triqs_vars, triqs_var)
                #exit()

                ### plot plot plot the stuff

                label = "G, max_time = " + str(max_time)
                plt.subplot(1, 3, 1)
                plt.title("$G_{triqs}-G_{pomerol}$")
                oplot(triqs['G'][s][b,b]-pom['G'][s][b,b], '-', x_window = (-iw_max,iw_max), mode = "R", name = label , color = color[nt])
                plt.plot(ymax,ymax, color="white")
                plt.plot(ymin,ymin, color="white")
                plt.ylabel(" ")

                plt.subplot(1, 3, 2)
                plt.title("$G_{w2dyn}-G_{pomerol}$")
                oplot(w2dyn['G'][s][b,b]-pom['G'][s][b,b], '-', x_window = (-iw_max,iw_max), mode = "R", name = label, color = color[nt])
                plt.plot(ymax,ymax, color="white")
                plt.plot(ymin,ymin, color="white")
                plt.ylabel(" ")

            print "spin = " + str(s) + ", orb_1 = " + str(b) + ", orb_2 = " + str(b)

            plt.subplot(1, 3, 3)
            plt.loglog(times, w2dyn_vars, "o-", label = "w2dyn")
            plt.loglog(times, triqs_vars, "o-", label = "triqs")
            plt.title("$var[G_{w2dyn}-G_{pomerol}]$")
            plt.ylabel(" ")

            plt.legend(loc = 'best')
            plt.show()
```
